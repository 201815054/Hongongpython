> Filter 함수, map 함수

```python
# 험수를 매개변수로 전달하는 대표적인 표준함수
# map(함수, 리스트)
# filter(함수, 리스트)

def power(item):
    return item * item 

def under_3(item): # 3보다 작니 물어봄 
    return item < 3 

li = [1,2,3,4,5]

# map() 함수를 이용
#      = map(power(item),[1,2,3,4,5] ) 
output = map(power, li)
print("map() 함수의 실행결과")
print(output)
# <map object at 0x0000020E987D5E40> 맵객체가 이렇게 된다는것. 
print(list(output)) #리스트를 만들어보면 맵객체를 가지고 리스트를 만든것. 제곱이나옴 
print()

# filter() 함수를 이용
print("filter() 함수의 실행결과")
output = filter(under_3, li)
print(output)
print(list(output))
# [1, 2]가나옴 트루인것만 필터링 된것. 
# 필터함수는 여기 꽂을때 함수자체가 어떤 기댓값이 트루폴스로 나오는 트루인경우 폴스인겨우 제외되는 필터링이 필요할때 필터를 가지고 실행결과를 맵핑시켜줌
```

```python
del(sum)

li = list(map(int, input().split()))
# print(sum([10,20,30]))

print(sum(li))
# 함수와 리스트를 맵핑시켜보면 인트로 형변환을 함 
```

```python
# 인라인 람다
li = [1, 2, 3, 4, 5]
output = map(lambda x: x ** 3 ,li)
print(list(output))

output = filter(lambda x: x <= 3, li)
print(list(output))
```

> csv 파일을 만들어 사용하기 

```python
# 텍스트를 사용해 데이터를 구조적으로 표현하는 방법 : csv, xml, json
# csv 파일 : 한 줄에 하나씩 ㅔㄷ이터를 나타매녀, 각가그이 줄은;,    (쉼표)로 데이터를 구분
# 김철수, 178 ,65

import random
first_names= "김이박최정강조윤장임한오서신"
last_names="가나다라마바사아자차카타파하"

with open("info.csv",'w') as f: 
    for i in range(10):
        first = random.choice(first_names)
        last = random.choice(last_names) + random.choice(last_names)
        name = first + last 

        height = random.randrange(140,200)
        weight = random.randrange(40,120) 
        f.write("{},{},{}".format(name,height,weight))
```

```python
# BMI(비만도) : 몸무게/키(m)의 제곱

with open("info.csv",'r', encoding='utf-8') as fo:
     for line in fo:
        (name,height,weight )= line.strip().split(', ')#튜플에 변수 자동으로 할당되게 하는것
        # 데이터가 유효한지 검사 
        if (not name) or (not weight) or (not height):
            continue #건너뛰기하는 키워드
        # BMI 계산
        bmi = int(weight) / int(height)/100 ** 2 #텍스트로 불러왔기때문에 int로 형변환시켜줘야함. #미터니까 /100 #** 2 제곱
        result = ""
        if bmi >= 25:
            result = "비만"
        elif bmi >= 23:
            result = "과체중"
        elif bmi >= 18:
            result = "정상"
        else:
            result ="저체중"
        
        str_out = '\n'.join([
            "이름:{}",
            "몸무게: {}",
            "키: {}", 
            "BMI: {}",
            "결과 :"]).format(name,weight,height,int(bmi),result)# 문자열을 만들어주어야 함. 
        print(str_out)
        print()
```
